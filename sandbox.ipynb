```python
import passiogo
#set the system to UNCC
system = passiogo.getSystemFromID(1053)
#Get routes
routes = system.getRoutes()
```

```python
for route in routes:
    print(f"ID: {route.id}, Name: {route.name} ({route.shortName})")

if routes:
    print(f"\n--- All data for the first route: '{routes[0].name}' ---")
    print(routes[0].__dict__)
```

```python
# Get a list of all stop objects
stops = system.getStops()

print(f"\nFound {len(stops)} total stops.")

# --- Play Option 1: Print all stop names ---
print("\n--- All Stop Names ---")
for stop in stops:
    print(f"ID: {stop.id}, Name: {stop.name}")

# --- Play Option 2: See all data for one stop ---
if stops:
    print(f"\n--- All data for the first stop: '{stops[0].name}' ---")
    print(stops[0].__dict__)
```
